# by the help of column transformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# Create a ColumnTransformer
transformer = ColumnTransformer(transformers=[
    # Transformer 1: Impute missing values in 'fever' column with the mean
    ('t1', SimpleImputer(), ['fever']),

    # Transformer 2: Ordinal encode 'cough' with custom order ['Mild', 'Strong']
    ('t2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']),

    # Transformer 3: One-hot encode 'gender' and 'city', drop first to avoid dummy variable trap
    ('t3', OneHotEncoder(sparse_output=False, drop='first'), ['gender', 'city'])
], remainder='passthrough')  # Leave all other columns unchanged
```

```python
transformer.fit_transform(x_train).shape
```
